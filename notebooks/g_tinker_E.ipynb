```python
# Running with Random Forest
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from random import choice
from random import randint
from random import random

from sklearn.linear_model import ElasticNet
from sklearn.linear_model import LinearRegression
df = pd.read_csv('../data/g1_merged.csv')
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
out = df

expl = df.select_dtypes(exclude=['object']).drop(['AFSeed', 'AUSeed','AFScore','AUScore','PFScore','PUScore'],axis=1)
rt = pd.DataFrame()
rr = 0
start = df.columns
```

```python
for py in range(2021,2022):
    # Targets and Predictors

    #py = choice([i for i in range(2008,2021) if i not in [2020]])
    df = df[start]
    fitX = expl[expl['Year']!=py]
    fitFY = df[df['Year']!=py]['PFScore']
    fitFU = df[df['Year']!=py]['PUScore']
    predX = expl[expl['Year']==py]

    # How well does Random Forest do?

    rs = randint(50,100)
    RFF = RandomForestRegressor(n_estimators=rs, oob_score=True, random_state=rs)
    RFF.fit(fitX,fitFY)
    RFU = RandomForestRegressor(n_estimators=rs, oob_score=True, random_state=rs)
    RFU.fit(fitX,fitFU)
    PF = RFF.predict(predX)
    PU = RFU.predict(predX)
    y = 0
    for x in out[out['Year']==py].index: 
        out.loc[x,'PFScore'] = PF[y]
        out.loc[x,'PUScore'] = PU[y]
        y = y + 1

    import numpy as np
    for x in out[out['Year']==py].index: 
        out.loc[x,'AWSeed'] = np.where(out.loc[x,'AFScore']>=out.loc[x,'AUScore'],out.loc[x,'AFSeed'],out.loc[x,'AUSeed'])
        out.loc[x,'AWTeam'] = np.where(out.loc[x,'AFScore']>=out.loc[x,'AUScore'],out.loc[x,'AFTeam'],out.loc[x,'AUTeam'])
        out.loc[x,'PWSeed'] = np.where(out.loc[x,'PFScore']>=out.loc[x,'PUScore'],out.loc[x,'PFSeed'],out.loc[x,'PUSeed'])
        out.loc[x,'PWTeam'] = np.where(out.loc[x,'PFScore']>=out.loc[x,'PUScore'],out.loc[x,'PFTeam'],out.loc[x,'PUTeam'])
        out.loc[x,'ESPN'] = np.where(out.loc[x,'AWTeam']==out.loc[x,'PWTeam'],10*2**(out.loc[x,'Round']-1),0)

    rft = out[out['Year']==py]['ESPN'].sum()

    # How well does Elastic Net do?

    alr = random()
    RFF = ElasticNet(alpha=alr,l1_ratio=alr)
    RFF.fit(fitX,fitFY)
    RFU = ElasticNet(alpha=alr,l1_ratio=alr)
    RFU.fit(fitX,fitFU)
    PF = RFF.predict(predX)
    PU = RFU.predict(predX)
    y = 0
    for x in out[out['Year']==py].index: 
        out.loc[x,'PFScore'] = PF[y]
        out.loc[x,'PUScore'] = PU[y]
        y = y + 1

    import numpy as np
    for x in out[out['Year']==py].index: 
        out.loc[x,'AWSeed'] = np.where(out.loc[x,'AFScore']>=out.loc[x,'AUScore'],out.loc[x,'AFSeed'],out.loc[x,'AUSeed'])
        out.loc[x,'AWTeam'] = np.where(out.loc[x,'AFScore']>=out.loc[x,'AUScore'],out.loc[x,'AFTeam'],out.loc[x,'AUTeam'])
        out.loc[x,'PWSeed'] = np.where(out.loc[x,'PFScore']>=out.loc[x,'PUScore'],out.loc[x,'PFSeed'],out.loc[x,'PUSeed'])
        out.loc[x,'PWTeam'] = np.where(out.loc[x,'PFScore']>=out.loc[x,'PUScore'],out.loc[x,'PFTeam'],out.loc[x,'PUTeam'])
        out.loc[x,'ESPN'] = np.where(out.loc[x,'AWTeam']==out.loc[x,'PWTeam'],10*2**(out.loc[x,'Round']-1),0)

    ent = out[out['Year']==py]['ESPN'].sum()

    # How well does Linear Regression do?


    RFF = LinearRegression()
    RFF.fit(fitX,fitFY)
    RFU = LinearRegression()
    RFU.fit(fitX,fitFU)
    PF = RFF.predict(predX)
    PU = RFU.predict(predX)
    y = 0
    for x in out[out['Year']==py].index: 
        out.loc[x,'PFScore'] = PF[y]
        out.loc[x,'PUScore'] = PU[y]
        y = y + 1

    import numpy as np
    for x in out[out['Year']==py].index: 
        out.loc[x,'AWSeed'] = np.where(out.loc[x,'AFScore']>=out.loc[x,'AUScore'],out.loc[x,'AFSeed'],out.loc[x,'AUSeed'])
        out.loc[x,'AWTeam'] = np.where(out.loc[x,'AFScore']>=out.loc[x,'AUScore'],out.loc[x,'AFTeam'],out.loc[x,'AUTeam'])
        out.loc[x,'PWSeed'] = np.where(out.loc[x,'PFScore']>=out.loc[x,'PUScore'],out.loc[x,'PFSeed'],out.loc[x,'PUSeed'])
        out.loc[x,'PWTeam'] = np.where(out.loc[x,'PFScore']>=out.loc[x,'PUScore'],out.loc[x,'PFTeam'],out.loc[x,'PUTeam'])
        out.loc[x,'ESPN'] = np.where(out.loc[x,'AWTeam']==out.loc[x,'PWTeam'],10*2**(out.loc[x,'Round']-1),0)

    lrt = out[out['Year']==py]['ESPN'].sum()
    rr = rr + 1
    rt.loc[rr,'Year']=py
    rt.loc[rr,'RF']=rft
    rt.loc[rr,'EN']= ent
    rt.loc[rr,'LR']=lrt
    rt
```
